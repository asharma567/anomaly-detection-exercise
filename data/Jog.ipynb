```python
rowapi = df['rows-loaded-api']
dayspans = [3,5,10]
plt.figure(figsize=(10,40))
for i,span in enumerate(dayspans):
    plt.subplot(3,1,i+1)
    rowapi.plot(color = color[0])
    rmean = pd.rolling_mean(rowapi,span)
    rmean.plot(label=span,color = color[i+3], alpha = .8)
    rstd = pd.rolling_std(rowapi,span)
    (rmean+2*rstd).plot(linestyle = '--', label=(str(span)+' +2 std-'+str(span)) ,color=color[3], alpha=0.5)
    (rmean-2*rstd).plot(linestyle = '--', label=(str(span)+' -2 std-'+str(span)) ,color=color[3], alpha=0.5)
    (rmean+3*rstd).plot(linestyle = ':', label=(str(span)+' +3 std-'+str(span)) ,color=color[4], alpha=0.5)
    (rmean-3*rstd).plot(linestyle = ':', label=(str(span)+' -3 std-'+str(span)) ,color=color[4], alpha=0.5)
    plt.legend(loc='upper left')
```

```python
pageviews = df['page-views']
# pageviews.plot()

plt.figure(figsize=(20,40))
for i,span in enumerate(dayspans):
    plt.subplot(3,1,i+1)
    pageviews.plot(color = color[0], linewidth =1)
    rmean = pd.rolling_mean(pageviews,span)
    rmean.plot(label='mean' + str(span),color = color[i+3], alpha = .8 )
    rstd = pd.rolling_std(pageviews,span)
    (rmean+2*rstd).plot(linestyle = '--', label=(str(span)+' +2 std-'+str(span)) ,color=color[3], alpha=0.5)
    (rmean-2*rstd).plot(linestyle = '--', label=(str(span)+' -2 std-'+str(span)) ,color=color[3], alpha=0.5)
    (rmean+3*rstd).plot(linestyle = ':', label=(str(span)+' +3 std-'+str(span)) ,color=color[4], alpha=0.5)
    (rmean-3*rstd).plot(linestyle = ':', label=(str(span)+' -3 std-'+str(span)) ,color=color[4], alpha=0.5)
    plt.legend(loc='upper left')
```
